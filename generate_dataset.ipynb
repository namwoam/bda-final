```python
import os
import glob
import pandas as pd
import numpy as np
from tqdm import tqdm
import random
```

```python
def list_subdirectory(directory, pattern='*'):
    # 使用 glob 模組列出資料夾中符合模式的檔案
    files = glob.glob(os.path.join(directory, pattern))
    return files
```

```python
parent_folder = "/home/namwoam/nas/homes/namwoam_data/bda-final/behavior"
target_folder = list_subdirectory(parent_folder)[0]
print(target_folder)
```

```python
files = list_subdirectory(target_folder)
print("csv file count:", len(files))
# files = files[:20]
print("sampled file count:", len(files))
```

```python
chunks = []
for file in tqdm(files, desc="Read csv files"):
    tmp_df = pd.read_csv(file)
    chunks.append(tmp_df)
df = pd.concat(chunks)
print("Dataset column count:", len(df.index))
df.head()
```

# Important: Sort actions by userID -> ShopID -> EventTime

```python
df = df.sort_values(by=['FullvisitorId', "ShopId", 'EventTime'])
df.head()
```

```python
df['EventTime'] = pd.to_datetime(df['EventTime'])
df = df.drop(columns=['Tunnel', 'Device', 'DeviceId', 'RegisterTunnel',
                      'SearchTerm', 'ContentType', 'ContentName', 'ContentId'])
df.head()
```

# Function to convert action data into sessions

```python
def calculate_time_to_checkout(df: pd.DataFrame):
    first_action_time = None
    current_user = None
    actions_in_interval = []
    intervals = []
    for _, row in tqdm(df.iterrows(), total=len(df.index),  desc="Parsing user behavior df"):
        if current_user is None or current_user != row['FullvisitorId']:
            current_user = row['FullvisitorId']
            actions_in_interval = []
            first_action_time = None
        if row['Behavior'] == 'purchase':
            if first_action_time:
                # record datapoint
                interval = row['EventTime'] - first_action_time
                intervals.append({
                    'customer_id': row['FullvisitorId'],
                    'actions': actions_in_interval.copy(),
                    'start_time': first_action_time,
                    'end_time': row['EventTime'],
                    'time_to_checkout': interval,
                    'trade_order': row['TradesGroupCode']
                })
            # renew first action
            first_action_time = row['EventTime']
            actions_in_interval = []
        else:
            if not first_action_time:
                first_action_time = row['EventTime']
            actions_in_interval.append(row['Behavior'])

    return pd.DataFrame(intervals)
```

```python
interval_df = calculate_time_to_checkout(df)
interval_df.head()
```

# Label users with hesitate / non-hesitate

```python
from datetime import timedelta


def post_process(df):
    df = df[df['time_to_checkout'] > timedelta(minutes=1)]
    mean_time = df['time_to_checkout'].mean()
    std_err = df["time_to_checkout"].std()
    df['hesitate'] = (df['time_to_checkout'] >
                      mean_time + 1*std_err).astype(int)
    return df
```

```python
result_df = post_process(interval_df)
result_df.head()
```

```python
print("All purchase actions:", len(result_df.index))
print("Hesitate purchase actions:", len(result_df[result_df["hesitate"] == 1]))
print("Percentage:", len(
    result_df[result_df["hesitate"] == 1])/len(result_df.index))
```
